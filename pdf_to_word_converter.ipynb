```python
from pdf2docx import Converter
import os

pdf_path = pdf_path  # uploaded PDF
output_docx = pdf_path.replace(".pdf", ".docx")

cv = Converter(pdf_path)

start = 0
end = len(cv.pages)
chunk_size = 50  # safe for large PDFs

for i in range(start, end, chunk_size):
    try:
        cv.convert(
            output_docx,
            start=i,
            end=min(i + chunk_size, end)
        )
        print(f"‚úÖ Converted pages {i} to {min(i + chunk_size, end)}")
    except Exception as e:
        print(f"‚ö†Ô∏è Skipped pages {i}-{i+chunk_size} due to error:", e)

cv.close()

print("üéØ Conversion finished (with safe skips if any)")
```

```python
import fitz
import pytesseract
from PIL import Image
from docx import Document

pdf = fitz.open(pdf_path)
doc = Document()

page_num = 358  # zero-based index

page = pdf[page_num]
pix = page.get_pixmap(dpi=300)
img_path = "failed_page.png"
pix.save(img_path)

text = pytesseract.image_to_string(Image.open(img_path), lang="eng")
text = text.replace('\x0c', '') # Remove form feed character
doc.add_paragraph(text)

doc.save("recovered_page_359.docx")
files.download(output_docx)
```
